# 可视化基础

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,50)
y = 2*x +1
y2 = x**2 +2

plt.figure()//这是第一张图
plt.plot(x,y)
plt.xlabel('This is x label')
plt.ylabel('This is y label')

plt.figure()// 这是第二张图
plt.plot(x,y2)

plt.show()

```

```python
plt.figure()
plt.plot(x,y,color ='r',linewidth =10,linestyle = '--',label = 'JYH')
plt.plot(x,y2,color = 'g',label='MNN')
plt.legend(loc='upper right')
plt.show()
```

# 鸢尾花可视化 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('D:\Data\iris\iris.csv')
print(data)
```

```python
flower1 = data[0:50]
flower2 = data[51:100]
flower3 = data[101:]
```

```python
plt.figure()
plt.scatter(flower1['Sepal.Length'],flower1['Sepal.Width'],color = 'r',label='flower1')
plt.scatter(flower2['Sepal.Length'],flower2['Sepal.Width'],color = 'g',label='flower2')
plt.scatter(flower3['Sepal.Length'],flower3['Sepal.Width'],color = 'b',label='flower3')
plt.show()
```

#  三维图绘制

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(flower1['Sepal.Length'],flower1['Sepal.Width'],flower1['Petal.Length'],c='r')
ax.scatter(flower2['Sepal.Length'],flower2['Sepal.Width'],flower2['Petal.Length'],c='g')
ax.scatter(flower3['Sepal.Length'],flower3['Sepal.Width'],flower3['Petal.Length'],c='b')
plt.show()
```
